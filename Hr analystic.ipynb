### Q.15) Identify departments with the highest attrition rate (Resigned %).

```python
df.head()
```

```python
pd=df.groupby("Department")["Status"].agg(total_emp = 'count', resigned = lambda x: (x == 'Resigned').sum())
pd
```

```python
type(pd)
```

```python
# Calculate attrition rate

pd['attrition_rate_%'] = (pd['resigned'] / pd['total_emp']) * 100
pd
```

```python
# Sort by attrition rate (highest first)
pd.sort_values("attrition_rate_%",ascending = False)
```
